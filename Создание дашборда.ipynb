### Исходные данные

Данные о конференциях TED: данные выступлений, справочник конференций, справочник авторов выступления.

### Задача

Исследовать историю TED-конференций с помощью Tableau.

### План исследования:

1. Изучим состав и типы данных.
2. Изучим, как количество конференций распределено по времени. Определим год, после которого количество выступлений скачкообразно выросло.
3. Рассмотрим  процент выступлений в каждой из стран.
4. Рассмотрим выступления по годам» с визуализацией.
5. Построим визуализацию процента выступлений по тематикам.
6. Создадим дашборд «История выступлений».
7. Создадим дашборд «Тематики выступлений», узнаем: 
    - Какие категории выступлений наиболее популярны
    - Различается ли распределение популярных категорий в разных странах? 
    - Какие категории чаще вызывают аплодисменты аудитории, а какие реже? Какому выступлению аплодировали больше остальных?
    - Какие категории чаще вызывают смех аудитории, а какие реже? Какое выступление оказалось самым смешным?
    - Есть ли зависимость между длительностью выступления и количеством просмотров? Какое выступление посмотрели чаще всего? Какое выступление длилось дольше всех?
8. Создадим дашборд «Авторы выступлений» и узнаем:
    - Какие области деятельности у авторов преобладают?
    - Сколько выступлений обычно приходится на одного автора? Кто выступал чаще всего?
    - Какой дизайнер (Designer) выступал чаще остальных? Какое выступление этого автора смотрели меньше всего?
9. Разделим конференции на категории.
10. Выясним как менялась популярность этих категорий со временем.
11. Подготовим презентацию.

Файлы tableau_project_data_1.csv, tableau_project_data_2.csv, tableau_project_data_3.csv хранят данные выступлений. У них одинаковая структура:
- talk_id — идентификатор выступления (целое число);
- url — ссылка на запись выступления(строка);
- title — название выступления (строка);
- description — краткое описание (строка);
- film_date — дата записи выступления(строка);
- duration — длительность в секундах(целое число);
- views — количество просмотров(целое число);
- main_tag — основная категория, к которой относится выступление (строка);
- speaker_id — уникальный идентификатор автора выступления(целое число);
- laughter_count — количество раз, когда аудитория смеялась в ходе выступления(число с плавающей точкой);
- applause_count — количество раз, когда аудитория аплодировала в ходе выступления(число с плавающей точкой);
- language — язык, на котором велось выступление(строка);
- event_id — уникальный идентификатор конференции(целое число).

### Основные выводы

- Больше всего выстпулений проходило в США, чуть больше половины от всех выступлений.
- К топ 5 самым популряным тематикам выстпулений относятся: global issues, society, culture, technology, science.
- К 2020 году колиечество выстпулений в топ-5 тем значительно уменьшилось, к 2021 году осталось по одному или вовсе не осталось выступлений с данными темами.
- Наиболее популряны выстпуления на темы science, technology, culture.
- Распределение категорий различается по странам. В Канаде выступления по теме science и technology популрянее, чем в США и UK. В Великобритании выстпуления на тему culture, business и design популярнее, чем США и Канаде.
- Аплодисменты чаще всего бывают в выступлениях по теме art и social change. Самые долгие аплодисменты были у выступления The political progress women have made -- and what's next по теме social change.
- Чаще всего смех у аудитории вызывают выступления по темам art и design. Самым смешным оказалось выступлеие A Saudi, an Indian and an Iranian walk into a Qatari bar из категории culture
- Между длительностью выступления и количеством просмотров зависимости нет.
- Больше всего просмотров (63 588 711) у выступления Your body language may shape who you are. Дольше всего (почти 63 минуты) длится выступление 3 secrets to Netflix's success.
- Наиболее популряные области деятельности у авторов: writer, artist, enterprenuer, journalist.
- На одного автора приходится от 1 до 8 выступлений, больше всего авторов с 1 выступлением. Автор, который выступал чаще всего - Hans Rosling, на его счету 8 выступлений.
- Чаще всего из дизайнеров выступал Tom Wujec, меньше всего из его выступлений смотрели Learn to use the 13th-century astrolabe.
- Конференции можно разбить на категории по названиям, самые попудярные конференции Tedx
- К 2019 году популярность всех категорий падает, конференции группы Tedx популрянее всех остальных на протяжении всего наблюдения, количество конференций в данной группе в разы больше, чем в остальных.
